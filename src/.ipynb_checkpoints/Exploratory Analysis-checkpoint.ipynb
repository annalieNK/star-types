# Exploratory Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('../data/prepared/data.pkl')
```

```python
print(df.shape)
df.head()
```

```python
df['Spectral Class'].value_counts()
```

```python
for i, (name, group) in enumerate(df.groupby('Spectral Class')):
    fig = plt.gcf()
    fig.set_size_inches(20,5)
    fig.suptitle('Temperature (K)', fontsize=24)
    ax = plt.subplot(1, 7, i+1)
    group['Temperature (K)'].hist()
    ax.set_title("{}".format(name))
    plt.ylabel("Frequency")
    plt.tight_layout();
```

```python
for i, (name, group) in enumerate(df.groupby('Spectral Class')):
    fig = plt.gcf()
    fig.set_size_inches(20,5)
    fig.suptitle('Luminosity(L/Lo)', fontsize=24)
    ax = plt.subplot(1, 7, i+1)
    group['Luminosity(L/Lo)'].hist()
    ax.set_title("{}".format(name))
    plt.ylabel("Frequency")
    plt.tight_layout();
```
